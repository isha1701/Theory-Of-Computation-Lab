PDA IMPLEMENTATION OF a^n b^m c^n where m,n>=1

```python
#test case 01
pda=dict()

pda[('q0','a')]='q0'
pda[('q0','b')]='q1'
pda[('q0','c')]='q4'
pda[('q0','z')]='q4'
pda[('q1','a')]='q4'
pda[('q1','b')]='q1'
pda[('q1','c')]='q2'
pda[('q1','z')]='q4'
pda[('q2','a')]='q4'
pda[('q2','b')]='q4'
pda[('q2','c')]='q2'
pda[('q2','z')]='q3'
pda[('q4','a')]='q4'
pda[('q4','b')]='q4'
pda[('q4','c')]='q4'
pda[('q4','z')]='q4'

class stack:
    def __init__ (self,pda):
        self.items=[]
    def push(self,item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def isempty(self):
        return self.items==[]
    def size(self):
        return len(self.items)
s=stack(pda)
x=input("Enter the string to be checked")
s.push('z')
cs='q0'
fs='q3'
for i in x:
    top=s.pop()
    s.push(top)
    cs=pda[cs,i]
    print("Current state is :",cs)
    if top=='z' and (i=='a' and cs=='q0'):
        s.push(i)
    elif top=='a' and (i=='a'and cs=='q0'):
        s.push(i)
    elif top=='a' and (i=='b'and cs=='q1'):
        continue
    elif top=='a' and (i=='c' and cs=='q2'):
        s.pop()
    elif top!='a' and (i=='b'or i=='c'):
        s.push(i)
y=s.pop()
cs=pda[cs,y]
if cs==fs:
    print("String Accepted")
else:
    print("String Not Accepted")
```
